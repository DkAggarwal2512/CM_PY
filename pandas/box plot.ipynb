```python
# Box Plot & Whisker Plot
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(7,3))
x =[60,70,80,90,99]
plt.boxplot(x)
#plt.show()
plt.savefig('box.png')
```

```python
plt.figure(figsize=(7,3))
x =[60,75,80,90,99,0]
plt.boxplot(x, vert=False, widths=0.4, labels=['Students Marks'], patch_artist=True, showmeans= True)
plt.show()
```

```python
plt.figure(figsize=(7,3))
x =[60,70,80,90,99]
plt.boxplot(x, notch =True)
plt.show()
```
