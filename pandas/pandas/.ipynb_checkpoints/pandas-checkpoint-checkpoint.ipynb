```python
import pandas as pd
df = pd.read_csv('sheet.csv')
df
```

```python
df.info()
```

```python
df.head(1)
```

```python
df['total'] = df['English']+df['Hindi']+df['Math']+df['Science']
```

```python
df
```

```python
df['percentage'] = df['total']/500*100
```

```python
df
```

```python
df.loc[(df.Hindi == 75)]
```

```python
df.loc[(df.English >= 55) & (df.Math >= 75)]
```

```python
df.loc[(df.English >= 65) | (df.Math >= 75)]
```

```python
df['Grade'] = 'Pass'
df
```

```python
#65+ distinction
#50-55 first
#45-50 pass
#30-40 fail

df.loc[df['percentage'] >= 65, ['Grade']] = 'Distinction'
df.loc[(df['percentage'] >=50) & (df['percentage'] <55), ['Grade']] = 'First'
df.loc[(df['percentage'] >=45)& (df['percentage'] <50), ['Grade']] = 'Pass'
df.loc[(df['percentage'] <40), ['Grade']] = 'Fail'
df.loc[~(df['percentage'] >=0), ['Grade']] = 'NaN'
df
```

```python
df.loc[df['Name'].str.contains('m')]
```
