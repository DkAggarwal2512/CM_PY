```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
df =pd.read_csv('marks1.csv')
df
```

```python
df=sns.load_dataset('tips')
df.head(1)
# df.tail(1)
```

```python
figr = sns.FacetGrid(df)
figr.map(plt.bar, 'total_bill', 'tip')
plt.show()
```

```python
figr = sns.FacetGrid(df, col='sex')
figr.map(plt.scatter, 'total_bill', 'tip')
plt.show()
```

```python
figr = sns.FacetGrid(df, col='day', hue='sex').add_legend()
figr.map(plt.scatter, 'total_bill', 'tip')
plt.show()
```

```python
figr = sns.FacetGrid(df, col='sex', hue='day', palette='cool').add_legend()
figr.map(plt.scatter, 'total_bill', 'tip')
```

```python
# pair plots
df.head(1)
```

```python
plt.figure(figsize=(4,4))
sns.pairplot(df, vars=['size', 'tip'],hue_order=['Male', 'Female'], y_vars=['tip'],palette='Accent', kind='hist')
plt.show()
```

```python
plt.figure(figsize=(4,4))
sns.pairplot(df, vars=['size', 'tip'],hue_order=['Male', 'Female'], y_vars=['tip'],palette='Accent', kind='kid')
plt.show()
```

```python
plt.figure(figsize=(4,4))
sns.pairplot(df, vars=['size', 'tip'],hue_order=['Male', 'Female'], y_vars=['tip'],palette='Accent', kind='scatter', markers=['D','<'])
plt.show()
```

```python
sns.catplot(x='total_bill', y ='tip', data=df)
```

```python
plt.figure(figsize=(4,4))
sns.catplot(y='size', x ='tip', data=df.head(30), hue='sex', palette='Accent', kind='strip')
```

```python
sns.catplot(y='size', x ='tip', data=df.head(30), hue='sex', palette='Accent', kind='point')
```

```python
sns.catplot(y='size', x ='tip', data=df.head(50), hue='sex', palette='Accent', kind='boxen')
```
