```python
def sig(a):
    import math
    return 1/(1+math.e**(-a))
```

```python
def pred_f(b,c,d):
    m = -7.20706471e-04*b+-9.59322438e-03*c+3.4273703e-06*d+(-0.00117166)
    n =sig(m)
    return n
```

```python
b =0
c =45
d = 150000
round(pred_f(b,c,d))
```
