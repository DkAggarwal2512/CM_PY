```python
import pandas as pd
```

```python
# data Mining
df= pd.read_csv('stu.csv')
df
```

```python
df.head(1)
```

```python
df.sample(1)
```

```python
x = df.drop('Salary', axis= 'columns')
x
```

```python
y = df['Salary']
y
```

```python
# labling for transform
from sklearn.preprocessing import LabelEncoder
```

```python
com_l = LabelEncoder()
job_l = LabelEncoder()
deg_l = LabelEncoder()
```

```python
x['com_1']= com_l.fit_transform(x['Company'])# indexing Elphabatic (0,2,1)
x
```

```python
x['job_1'] = job_l.fit_transform(x['Job'])
```

```python
x['deg_1'] = deg_l.fit_transform(x['Degree'])
```

```python
y.replace({'<50000':0, '>50000':1}, inplace= True)
y
```

```python
x1 = x.drop(['Company','Job','Degree'], axis='columns')
x1
```

```python
x1.head(1)
       
```

```python
x1['total']='0'
x1
```

```python
# Decision Tree
from sklearn import tree
```

```python
model =tree.DecisionTreeClassifier()
model.fit(x1,y)
```

```python
model.score(x1,y)
```

```python
model.predict([[2,1,0,1]])
```
