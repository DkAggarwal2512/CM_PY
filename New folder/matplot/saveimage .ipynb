# pairplot

```python
x = [ 45,65,75,85]
y = [ 'Republican ', ' Democrats','Orthodox','Conservation']
ex =[0.3,0.0,0.0,0.0]
color=['r','y','b','g']
plt.pie(x, labels =y, explode=ex, colors=color)
plt.show()
```

```python
x = [ 45,65,75,85]
y = [ 'Republican ', ' Democrats','Orthodox','Conservation']
ex =[0.3,0.0,0.0,0.0]
plt.pie(x, labels =y, explode=ex, autopct= "%0.2f%%", shadow=True)
plt.show()
```

```python
x = [ 45,65,75,85]
y = [ 'Republican ', ' Democrats','Orthodox','Conservation']
ex =[0.3,0.0,0.0,0.0]
plt.pie(x, labels =y, explode=ex, autopct= "%0.2f%%", radius=0.50)
plt.show()
```

```python
x = [ 45,65,75,85]
y = [ 'Republican ', ' Democrats','Orthodox','Conservation']
ex =[0.3,0.0,0.0,0.0]
plt.pie(x, labels =y, explode=ex, autopct= "%0.2f%%", startangle=180,
       labeldistance=1,shadow= True,counterclock= False,
       rotatelabels= True)
plt.legend(loc= 'lower right')
plt.savefig('fig.png')
plt.show()
```
