# # 9 oct ..  Bar plot in Matplotlib

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('ggg.csv')
df
```

```python
plt.plot(df['Name'],df['English'])
plt.show()
```

```python
import numpy as np
plt.figure(figsize=(7,3))


x = np.array(['Allahabad','Dehradun','Kanpur','Patna'])
y = np.array([17,21,17,10])
plt.bar(x,y)
plt.show()
```

```python
plt.figure(figsize=(7,3))
plt.bar(df['Name'],df['Maths'])
plt.show()
```

```python
#cmykrgbw -> cyan , magenta,yello,black,red,green,blue,,white
#size 
plt.figure(figsize=(7,3))

color =['r','m','c','g']
plt.bar(df['Name'],df['Maths'],color=color)



#title
plt.title("Student's Marks", fontsize=15)
plt.xlabel('Students')
plt.ylabel('Marks')

plt.show()
```

```python
#cmykrgbw -> cyan , magenta,yello,black,red,green,blue,,white
#size 
plt.figure(figsize=(7,3))

color =['r','m','c','g']
plt.bar(df['Name'],df['Maths'],color=color,width=0.4,align ='center',
edgecolor='b',linewidth=5,linestyle='dashed')



#title
plt.title("Student's Marks", fontsize=15)
plt.xlabel('Students')
plt.ylabel('Marks')

plt.show()
```

```python
#cmykrgbw -> cyan , magenta,yello,black,red,green,blue,,white
#size 
plt.figure(figsize=(7,3))

color =['r','m','c','g']
plt.bar(df['Name'],df['Maths'],color=color,width=0.4,align ='center', label='annnual exam result ')



#title
plt.title("Student's Marks", fontsize=15)
plt.xlabel('Students')
plt.ylabel('Marks')


plt.legend()
plt.show()
```

```python
plt.figure(figsize=(7,3))
plt.barh(df['Name'],df['Maths'])
plt.show()
```
