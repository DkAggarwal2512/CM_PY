# 17 oct   catplot, paiplot,facet Grid

```python

import matplotlib.pyplot as plt
import pandas as pd 
import seaborn as sns
```

```python
df = sns.load_dataset('tips')
```

```python
df.head(1)
```

```python
figer = sns.FacetGrid(df)
figer.map(plt.scatter,'total_bill','tip')
plt.show()
```

```python
figer = sns.FacetGrid(df, col='day')
figer.map(plt.scatter,'total_bill','tip')
plt.show()
```

```python
figer = sns.FacetGrid(df, col='day')
figer.map(plt.bar,'smoker','tip')
plt.show()
```

```python
figer = sns.FacetGrid(df, col='day', hue='sex')
figer.add_legend()
figer.map(plt.scatter,'smoker','tip')
plt.show()
```

```python
figer = sns.FacetGrid(df, col='day', hue='day', palette='cool')
figer.map(plt.scatter,'smoker','tip')
plt.show()
```
