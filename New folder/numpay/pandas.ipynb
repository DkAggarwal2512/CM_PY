```python
a['total ']=a['English']+a['Hindi']+a['Maths']+a['Science']+a['Social Science']
```

```python

a
```

```python
a.loc[(a['English'] >=90)  & (a['Maths'] >=75)]
```

```python
a['Percentage']=(a['total']/500)* 100
```

```python
a.loc[(a['English'] >=90)  |  (a['Maths'] >=95)]
```

```python
a.loc[(a['English'] >=90)  & (a['Maths'] >=75) | (a['Science']>=80)]
```

```python
a['Grade']= 'Pass'
```

```python
a
```

```python
# 85 +  ->  Distinction
# 80-85 -> first
# 75-80 -> pass
# 70-75 -> fail


a.loc[a['Percentage'] >=85 , ['Grade']] =' Distinction'
a.loc[(a['Percentage'] >=80) & (a['Percentage']<85), ['Grade']]= 'First'
a.loc[(a['Percentage'] >=75) & (a['Percentage']<80), ['Grade']]= 'Pass'
a.loc[a['Percentage'] <=75 , ['Grade']]= 'Fail'
```
