```python
def gradie(x,y):
    m_curr = c_curr = 0
    rate = 0.01
    n = len(x)
    plt.scatter(x, y, marker='*', linewidth=2)

    for i in range(10000):
        y_pred = m_curr * x + c_curr
        plt.plot(x, y_pred, color='g')
        md = -(2/n) * sum(x * (y - y_pred))
        cd = -(2/n) * sum(y - y_pred)
        m_curr = m_curr - rate * md
        c_curr = c_curr - rate * cd
```

```python
# x = [1, 2, 3, 4, 5]
# y = [2, 3, 2.5, 4, 3.5]
x= ['Age']
y=['Income']
gradie(x, y)
```
