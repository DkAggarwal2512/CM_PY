```python
from sklearn.neighbors import KNeighborsClassifier
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('data.csv')
```

```python
df
```

```python
x= df.drop('Insurance',axis='columns')
```

```python
y=df['Insurance']
```

```python
model = KNeighborsClassifier(n_neighbors=5)
```

```python
model.fit(x,y)
```

```python
model.predict([[22,6300]])
```

```python
df['pred']=model.predict(x)
```

```python
df
```

```python
from sklearn.metrics import confusion_matrix
```

```python
cm =confusion_matrix(df['Insurance'], df['pred'])
cm 
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', linewidths=0.5, linecolor='black')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Age', y='Income', data=df, hue='Insurance')
plt.xlabel('Age')
plt.ylabel('Income')
plt.show()
```
