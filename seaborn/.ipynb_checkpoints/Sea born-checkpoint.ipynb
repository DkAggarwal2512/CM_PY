```python
import mpl_toolkits as mpt
import numpy as np
import matplotlib.pyplot as plt
```

```python
#plot3D
x = np.array([1,2,3,4,5])
y = np.array([3,5,1,9,6])
z = np.ones(5)
```

```python
fig = plt.figure()
ax = plt.axes(projection ='3d')
ax.plot3D(x,y,z, marker='D',color='y',mec='k')
plt.show()
```

```python
#Barplot
x = np.array([1,2,3,4,5,6,7])
y = np.array([3,5,1,9,6,4,5])
z = np.ones(7)
```

```python
xsize=np.ones(7)
ysize=np.ones(7)
zsize=np.array([2,3,2,5,3,4,2])
```

```python
fig = plt.figure()
ax = plt.axes(projection ='3d')
ax.bar3d(x,y,z,xsize,ysize,zsize)
plt.show()
```

```python
#Scatter plot
x = np.array([1,2,3,4,5,6,7])
y = np.array([3,5,1,9,6,4,5])
z = np.ones(7)
```

```python
fig = plt.figure()
ax = plt.axes(projection ='3d')
ax.scatter3D(x,y,z, c='m')
plt.show()
```
